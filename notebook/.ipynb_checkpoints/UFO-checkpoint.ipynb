```python
def get_date(s):
    d = s.split(' ')[0]
    d = pd.to_datetime(d)
    return d
```

```python
def get_hour(s):
    h = s.split(' ')[-1].split(':')[0]
    return h
```

```python
#get_hour(s)
```

```python
def get_time(s):
    t = s.split(' ')[-1]
    return t
```

```python
#get_time(s)
```

```python
df['date_pd'] = df['datetime'].apply(get_date)
```

```python
df['time_pd'] = df['datetime'].apply(get_time)
```

## Clustering

```python
df3 = pd.read_csv('../data/full_final_UFO.csv')
```

```python
df3.columns.values
```

```python
cols = ['state', 'shape', 'duration (seconds)', 'phaseid', 'Month_sighting','Extreme_Weather', 'hour_sighting', 
       'dow_sighting', 'Proximity to USAFB']
df_clus = df3[cols]
```

```python
df_clus.head()
```

```python
cat_cols = ['state', 'shape',  'phaseid', 'Month_sighting','Extreme_Weather', 'hour_sighting', 
       'dow_sighting']
df_clus_dummy = pd.get_dummies(df_clus, columns=cat_cols)
```

```python
df_clus_dummy.shape
```

```python
df_clus_dummy.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df_clus_dummy)
```

```python
df_clus_dummy = scaler.transform(df_clus_dummy)
```

```python
df_clus_dummy[:10]
```

```python
from sklearn.decomposition import PCA
```

```python
pca_80 = PCA(0.80)
```

```python
pca_80.fit(df_clus_dummy)
```

```python
df_clus_dummy_pca = pca_80.transform(df_clus_dummy)
```

```python
df_clus_dummy_pca
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(init="k-means++", n_clusters=8)
```

```python
kmeans.fit(df_clus_dummy_pca)
```

```python
#kmeans.cluster_centers_
```

```python
labels = kmeans.predict(df_clus_dummy_pca)
```

```python
centroids = kmeans.cluster_centers_
```

```python
labels
```

```python
np.unique(labels)
```

```python
labels.shape
```

```python
df_clus['Cluster_No'] = labels
```
