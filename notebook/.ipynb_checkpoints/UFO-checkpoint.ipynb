```python
def get_date(s):
    d = s.split(' ')[0]
    d = pd.to_datetime(d)
    return d
```

```python
def get_hour(s):
    h = s.split(' ')[-1].split(':')[0]
    return h
```

```python
#get_hour(s)
```

```python
def get_time(s):
    t = s.split(' ')[-1]
    return t
```

```python
#get_time(s)
```

```python
df['time_pd'] = df['datetime'].apply(get_time)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_clus_dummy = scaler.transform(df_clus_dummy)
```

```python
df_clus_dummy[:10]
```

```python
from sklearn.decomposition import PCA
```

```python
pca_80 = PCA(0.80)
```

```python
pca_80.fit(df_clus_dummy)
```

```python
df_clus_dummy_pca = pca_80.transform(df_clus_dummy)
```

## Week number of the year

```python
import datetime
testdate=datetime.datetime(2010,6,16)
```

```python
def week_Number(date):
    testdate = pd.to_datetime(date)
    wn = ((testdate - datetime.datetime(testdate.year,1,1)).days // 7) + 1
    return wn
```

```python
df3['week_number'] = df3['date_pd'].map(week_Number)
```

```python
df3.head(3)
```
