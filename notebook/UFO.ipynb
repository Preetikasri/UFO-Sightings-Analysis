```python
def get_date(s):
    d = s.split(' ')[0]
    d = pd.to_datetime(d)
    return d
```

```python
get_date(s)
```

```python
def get_hour(s):
    h = s.split(' ')[-1].split(':')[0]
    return h
```

```python
get_hour(s)
```

```python
def get_time(s):
    t = s.split(' ')[-1]
    return t
```

```python
get_time(s)
```

```python
df['date_pd'] = df['datetime'].apply(get_date)
```

```python
df['time_pd'] = df['datetime'].apply(get_time)
```
