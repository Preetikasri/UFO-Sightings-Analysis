```python
import pandas as pd
import numpy as np
```

```python
moon = pd.read_csv('C:\Users\cathe\Documents\Python docs\UFO_sightings\UFO-Sightings-Analysis\data\moon_phase_data.csv') #read in moon phases
weather = pd.read_csv('C:\Users\cathe\Documents\Python docs\UFO_sightings\UFO-Sightings-Analysis\data\weather.csv')
```

```python
weather['STATE'] = weather['STATE'].str.lower() #make states lowercase to match the other data
weather = weather.drop_duplicates(['STATE','Date'])
weather[:25]
```

```python
ufo_atmosphere = pd.merge(moon, weather, left_on=[moon.state, moon.date], right_on=[weather.STATE, weather.Date], how='left')
```

```python
import os
os.chdir('C:\Users\cathe\Documents\Python docs\UFO_sightings\UFO-Sightings-Analysis\data')
ufo_atmosphere.to_csv('extreme_weather.csv',sep=',', encoding='utf-8')
```
